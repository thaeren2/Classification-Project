```python
"""
==============================
الجزء الأول: استيراد المكتبات
==============================
"""
import torch
import torch.nn as nn
import torch.optim as optim
import gymnasium as gym
import numpy as np
from collections import deque
import random
import matplotlib.pyplot as plt

"""
==============================
الجزء الثاني: تعريف الشبكة العصبية
==============================
هذه هي الدماغ التي ستتعلم اللعب
"""
class DQN(nn.Module):
    def __init__(self, state_dim, action_dim):
        super(DQN, self).__init__()
        # طبقات الشبكة العصبية
        self.fc1 = nn.Linear(state_dim, 128)  # طبقة أولى: 128 خلية
        self.fc2 = nn.Linear(128, 128)         # طبقة ثانية: 128 خلية
        self.fc3 = nn.Linear(128, action_dim)  # طبقة مخرجات: عدد الإجراءات

    def forward(self, x):
        # تمرير البيانات عبر الشبكة
        x = torch.relu(self.fc1(x))  # تنشيط ReLU
        x = torch.relu(self.fc2(x))
        return self.fc3(x)

"""
==============================
الجزء الثالث: تعريف وكيل DQN
==============================
هذا هو الوكيل الذي سيتعلم
"""
class DQNAgent:
    def __init__(self, state_dim, action_dim):
        # المعاملات الأساسية
        self.state_dim = state_dim
        self.action_dim = action_dim
        self.gamma = 0.99      # عامل الخصم (كم نهتم بالمكافآت المستقبلية)
        self.epsilon = 1.0     # معدل الاستكشاف (100% استكشاف في البداية)
        self.epsilon_min = 0.01
        self.epsilon_decay = 0.995
        self.batch_size = 64
        self.memory = deque(maxlen=10000)  # ذاكرة الخبرات

        # إنشاء الشبكتين
        self.q_network = DQN(state_dim, action_dim)
        self.target_network = DQN(state_dim, action_dim)
        self.target_network.load_state_dict(self.q_network.state_dict())

        # المحسن (Optimizer)
        self.optimizer = optim.Adam(self.q_network.parameters(), lr=0.001)
        self.step_count = 0

    """
    اختيار الإجراء: إما استكشاف (عشوائي) أو استغلال (أفضل إجراء)
    """
    def act(self, state):
        if np.random.rand() < self.epsilon:
            return np.random.randint(self.action_dim)  # استكشاف
        state_tensor = torch.FloatTensor(state).unsqueeze(0)
        q_values = self.q_network(state_tensor)
        return q_values.argmax().item()  # استغلال

    """
    تخزين الخبرة في الذاكرة
    """
    def remember(self, state, action, reward, next_state, done):
        self.memory.append((state, action, reward, next_state, done))

    """
    تحديث الشبكة باستخدام عينات عشوائية من الذاكرة
    """
    def replay(self):
        if len(self.memory) < self.batch_size:
            return

        batch = random.sample(self.memory, self.batch_size)
        states = torch.FloatTensor([b[0] for b in batch])
        actions = torch.LongTensor([b[1] for b in batch])
        rewards = torch.FloatTensor([b[2] for b in batch])
        next_states = torch.FloatTensor([b[3] for b in batch])
        dones = torch.FloatTensor([b[4] for b in batch])

        # حساب القيمة الحالية
        current_q = self.q_network(states).gather(1, actions.unsqueeze(1)).squeeze()

        # حساب القيمة المستهدفة
        next_q = self.target_network(next_states).max(1)[0]
        target_q = rewards + self.gamma * next_q * (1 - dones)

        # حساب الخسارة وتحديث الأوزان
        loss = nn.MSELoss()(current_q, target_q.detach())
        self.optimizer.zero_grad()
        loss.backward()
        self.optimizer.step()

        # تحديث الشبكة الهدف كل 10 حلقات
        self.step_count += 1
        if self.step_count % 10 == 0:
            self.target_network.load_state_dict(self.q_network.state_dict())

        return loss.item()

    """
    تقليل معدل الاستكشاف تدريجياً
    """
    def decay_epsilon(self):
        self.epsilon = max(self.epsilon_min, self.epsilon * self.epsilon_decay)

"""
==============================
الجزء الرابع: تدريب الوكيل
==============================
"""
def train_dqn(episodes=300):
    # إنشاء البيئة (لعبة CartPole)
    env = gym.make('CartPole-v1')
    state_dim = env.observation_space.shape[0]
    action_dim = env.action_space.n

    # إنشاء الوكيل
    agent = DQNAgent(state_dim, action_dim)

    # تخزين النتائج
    scores = []
    losses = []

    print("بدء التدريب...")

    for episode in range(episodes):
        state, _ = env.reset()
        score = 0
        episode_loss = []

        while True:
            # اختيار إجراء
            action = agent.act(state)

            # تنفيذ الإجراء في البيئة
            next_state, reward, done, truncated, _ = env.step(action)
            done = done or truncated

            # تخزين الخبرة
            agent.remember(state, action, reward, next_state, done)

            # تحديث الشبكة
            loss = agent.replay()
            if loss:
                episode_loss.append(loss)

            state = next_state
            score += reward

            if done:
                break

        # تقليل الاستكشاف بعد كل حلقة
        agent.decay_epsilon()
        scores.append(score)
        if episode_loss:
            losses.append(np.mean(episode_loss))

        # عرض التقدم
        if episode % 50 == 0:
            print(f"الحلقة {episode}: النقاط = {score}, معدل الاستكشاف = {agent.epsilon:.3f}")

    env.close()
    print("تم التدريب!")
    return scores, losses

"""
==============================
الجزء الخامس: تشغيل التدريب ورسم النتائج
==============================
"""
if __name__ == "__main__":
    print("تشغيل التدريب...")
    scores, losses = train_dqn(episodes=300)

    # رسم النتائج
    plt.figure(figsize=(14, 5))

    # رسم النقاط
    plt.subplot(1, 2, 1)
    plt.plot(scores)
    plt.title('نتائج التدريب')
    plt.xlabel('عدد الحلقات')
    plt.ylabel('النقاط')
    plt.grid(True)

    # رسم الخسارة
    plt.subplot(1, 2, 2)
    plt.plot(losses)
    plt.title('خسارة التدريب')
    plt.xlabel('عدد الحلقات')
    plt.ylabel('الخسارة')
    plt.grid(True)

    plt.tight_layout()
    plt.savefig('نتائج_التدريب.png')
    plt.show()
    print("تم حفظ الرسم البياني باسم 'نتائج_التدريب.png'")
```
